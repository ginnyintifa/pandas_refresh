### Task 5: Handle Missing Data
A DataFrame with `NaN` values has been created for you.
1. Find which cells are null using `.isnull()`
2. Check how many nulls are in each column using `.isnull().sum()`
3. Drop rows where `weight` is missing using `.dropna(subset=)`
4. Fill missing `height` values with the median height using `.fillna()`

```python
# Starter data — don't change this
players = pd.DataFrame({
    'name':   ['Alice', 'Bob', 'Carol', 'Dave', 'Emma'],
    'height': [165, np.nan, 172, 180, np.nan],
    'weight': [60, 75, np.nan, 85, 62],
    'score':  [88, np.nan, 91, 74, 95],
})

# 1. Show which values are null
players.isnull()

# 2. Count nulls per column
players.isnull().sum()

# 3. Drop rows where weight is missing
players.dropna(subset='weight')

# 4. Fill missing height with the median
players['height']= players['height'].fillna(players['height'].median())
players
```

### Task 6: Rename & Drop Columns
A DataFrame has been created for you.
1. Rename the column `qty` to `quantity` and `amt` to `amount` using `.rename()`
2. Drop the `internal_code` column using `.drop()`
3. Sort by `amount` descending

```python
# Starter data — don't change this
inventory = pd.DataFrame({
    'product':       ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'qty':           [10, 150, 80, 25, 60],
    'amt':           [999.99, 29.99, 79.99, 349.99, 199.99],
    'internal_code': ['A001', 'A002', 'A003', 'A004', 'A005'],
})

# 1. Rename qty -> quantity, amt -> amount

inventory.rename(columns={'qty':'quantity','amt':'amount'},inplace=True)
# 2. Drop the internal_code column
inventory.drop(columns='internal_code',inplace = True)

# 3. Sort by amount descending
inventory.sort_values('amount',ascending=False)
```

### Task 7: GroupBy with Multiple Aggregations
Using the `scores` DataFrame below, group by `subject` and calculate the **mean, min, and max** of `score` in a single step using `.agg()`.

```python
# Starter data — don't change this
scores = pd.DataFrame({
    'student': ['Alice', 'Bob', 'Carol', 'Dave', 'Emma', 'Alice', 'Bob', 'Carol'],
    'subject': ['Math', 'Math', 'Math', 'Math', 'Science', 'Science', 'Science', 'Science'],
    'score':   [85, 92, 78, 88, 90, 76, 84, 95],
})

# Group by subject and calculate mean, min, max of score
scores.groupby('subject')['score'].agg(['mean','min','max'])
```

### Task 8: Pivot Table
Using the same `scores` DataFrame, create a pivot table showing the **average score** for each combination of `student` and `subject`.

```python
# Create a pivot table

pd.pivot_table(scores,values='score', columns='student',index='subject', aggfunc = 'mean')
```

### Task 10: DateTime Operations
A DataFrame of transactions has been created for you. Using the `.dt` accessor:
1. Extract the **month** from the `date` column into a new column `month`
2. Extract the **day of the week** (0=Monday, 6=Sunday) into a new column `weekday`
3. Find the **total amount spent per month** using `.groupby('month')`
4. Find which **weekday** has the highest average spending

```python
# Starter data — don't change this
np.random.seed(42)
dates = pd.date_range(start='2025-01-01', end='2025-12-31', freq='D')
txns = pd.DataFrame({
    'date':   np.random.choice(dates, size=200),
    'amount': np.random.uniform(5, 300, size=200).round(2),
})

# 1. Extract month
txns['month'] = txns['date'].dt.month

# 2. Extract day of week

txns['weekday'] = txns['date'].dt.weekday

# 3. Total amount per month
txns.groupby('month')['amount'].sum()

# 4. Average spending per weekday
txns.groupby('weekday')['amount'].mean()
```
