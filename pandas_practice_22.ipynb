### Task 2: .pipe() — write your own function
A retail products DataFrame has been created for you, along with two helper functions.

Your job:
1. **Write a third function yourself** — `add_discount(df)` — that adds a `discounted_price` column: `price * 0.9` if `stock > 50`, else `price * 0.95`.
   - Use `np.where` inside the function.
2. Use `.pipe()` to apply all three in order: `add_margin`, `flag_low_stock`, then your `add_discount`.
3. How many items are flagged as low stock?
4. What is the mean `margin`? Use NumPy.
5. Which `category` has the highest mean `discounted_price`? Use groupby and `.idxmax()`.

```python
# Starter data — don't change this
np.random.seed(5)
products = pd.DataFrame({
    'product_id': [f'P{i:03d}' for i in range(1, 9)],
    'category':   np.random.choice(['Electronics', 'Clothing', 'Home'], size=8),
    'price':      np.round(np.random.uniform(10, 200, size=8), 2),
    'cost':       np.round(np.random.uniform(5, 100, size=8), 2),
    'stock':      np.random.randint(5, 120, size=8),
})

def add_margin(df):
    df['margin'] = df['price'] - df['cost']
    return df

def flag_low_stock(df, threshold=30):
    df['low_stock'] = df['stock'] < threshold
    return df

# Your code here — define add_discount, then use .pipe()

def add_discount(df):
    df['discounted_price'] = np.where(df['stock']>50, df['price']*0.9, df['price']*0.95)
    return df

res = (
    products
    .pipe(add_margin)
    .pipe(flag_low_stock)
    .pipe(add_discount)
)

res['low_stock'].sum()
np.mean(res['margin'])
res.groupby('category')['discounted_price'].mean().idxmax()
```

### Task 5: pd.merge() + pd.concat() + pivot_table
Two DataFrames have been created for you: `warehouses` and `shipments`.

1. Inner join `shipments` with `warehouses` on `warehouse_id`.
2. Add `region_avg_weight`: mean `weight_kg` per `region` using `transform`.
3. Add `heavy`: `True` if `weight_kg > region_avg_weight`.
4. Two separate order DataFrames (`orders_q1`, `orders_q2`) have also been created. Concatenate vertically, reset the index. Store as `orders`.
5. Pivot `orders`: mean `amount` by `region` (rows) and `quarter` (columns).

```python
# Starter data — don't change this
np.random.seed(62)
warehouses = pd.DataFrame({
    'warehouse_id': [f'W{i:02d}' for i in range(1, 7)],
    'region':       np.random.choice(['North', 'South', 'West'], size=6),
    'capacity':     np.random.randint(500, 2000, size=6),
})

shipments = pd.DataFrame({
    'shipment_id':  [f'SH{i:03d}' for i in range(1, 25)],
    'warehouse_id': np.random.choice([f'W{i:02d}' for i in range(1, 7)], size=24),
    'weight_kg':    np.round(np.random.uniform(1, 50, size=24), 1),
    'quarter':      np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], size=24),
})

orders_q1 = pd.DataFrame({
    'warehouse_id': np.random.choice([f'W{i:02d}' for i in range(1, 7)], size=8),
    'region':       np.random.choice(['North', 'South', 'West'], size=8),
    'amount':       np.random.randint(100, 1000, size=8),
    'quarter':      'Q1',
})

orders_q2 = pd.DataFrame({
    'warehouse_id': np.random.choice([f'W{i:02d}' for i in range(1, 7)], size=8),
    'region':       np.random.choice(['North', 'South', 'West'], size=8),
    'amount':       np.random.randint(100, 1000, size=8),
    'quarter':      'Q2',
})

# Your code here
ij = pd.merge(
    warehouses, 
    shipments,
    how='inner',
    on='warehouse_id'
)
ij['region_avg_weight'] = ij.groupby('region')['weight_kg'].transform('mean')
ij['heavy'] = ij['weight_kg']>ij['region_avg_weight']
orders = pd.concat(
    [orders_q1, orders_q2],
    ignore_index=True
)
p = orders.pivot_table(
    index='region',
    columns='quarter',
    values='amount'
)


p





```

### Task 6: Full Pipeline
Three DataFrames have been created for you: `products2`, `stores2`, and `sales2`.
Two helper functions have been defined. **Write the third yourself.**

1. Inner join `sales2` with `stores2` on `store_id`. Inner join with `products2` on `product_id`. Drop nulls.
2. **Write a function `add_profit_margin(df)`** that adds a `profit_margin` column: `(revenue - total_cost) / revenue`, rounded to 2 decimal places.
   - `revenue` = `unit_price * qty`
   - `total_cost` = `unit_cost * qty`
3. Use `.pipe()` to apply all three in order: `add_revenue`, your `add_profit_margin`, then `flag_high_value`.
4. Add `priority` (row-wise): `'High'` if `qty >= 6` **and** `high_value` is `True`, `'Medium'` if either is true, else `'Low'`.
5. Pivot mean `revenue` by `city` (rows) and `category` (columns). Stack the result.
6. Use `.xs()` to extract just `'Furniture'` from the stacked result.
7. Find the correlation between `profit_margin` and `qty` using NumPy.

```python
# Starter data — don't change this
np.random.seed(77)

products2 = pd.DataFrame({
    'product_id': [f'PR{i:02d}' for i in range(1, 7)],
    'name':       ['Desk Chair', 'Standing Desk', 'Monitor', 'Keyboard', 'Bookshelf', 'Lamp'],
    'category':   ['Furniture', 'Furniture', 'Electronics', 'Electronics', 'Furniture', 'Electronics'],
    'unit_price':  [120, 350, 250, 80, 200, 45],
    'unit_cost':   [60, 180, 130, 35, 90, 20],
})

stores2 = pd.DataFrame({
    'store_id': [f'ST{i:02d}' for i in range(1, 7)],
    'city':     np.random.choice(['Portland', 'Houston', 'Denver', 'Miami'], size=6),
    'tier':     np.random.choice(['A', 'B', 'C'], size=6),
})

sales2 = pd.DataFrame({
    'sale_id':    [f'S{i:04d}' for i in range(1, 25)],
    'store_id':   np.random.choice([f'ST{i:02d}' for i in range(1, 7)], size=24),
    'product_id': np.random.choice([f'PR{i:02d}' for i in range(1, 7)], size=24),
    'qty':        np.random.randint(1, 10, size=24),
    'sale_date':  pd.date_range('2024-01-01', periods=24, freq='12D').strftime('%Y-%m-%d'),
})

def add_revenue(df):
    df['revenue'] = df['unit_price'] * df['qty']
    return df

def flag_high_value(df, threshold=500):
    df['high_value'] = df['revenue'] > threshold
    return df

# Your code here — define add_profit_margin, then build the pipeline
ij = pd.merge(
    sales2,
    stores2,
    on='store_id',
    how='inner'
)
ij2 = pd.merge(
    ij, 
    products2,
    on='product_id',
    how = 'inner'
).dropna()

def add_profit_margin(df):
    df['profit_margin'] = round((df['revenue']-df['unit_cost']*df['qty'])/(df['revenue']),2)
    return df
res = (
    ij2
    .pipe(add_revenue)
    .pipe(add_profit_margin)
    .pipe(flag_high_value)
)
res['priority'] = res.apply(lambda row: 'High' if row['qty']>=6 and row['high_value']
                            else 'Medium' if row['qty']>=6 or row['high_value']
                            else 'Low',
                            axis = 1)
p = res.pivot_table(
    index='city',
    columns='category',
    values = 'revenue'
).stack()
p.xs('Furniture',level='category')
np.corrcoef(res['profit_margin'],res['qty'])
```
