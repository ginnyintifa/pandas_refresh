## Level 1 — Flights

Use `sns.load_dataset('flights')`. Group by `year`.

Build a named-agg summary with `total` (sum of `passengers`), `peak` (max), `low` (min), and `seasonal_range` (max minus min via lambda).

- Which year had the most total passengers? Which had the widest seasonal range?
- Rank years from widest to narrowest seasonal range using `np.argsort`.
- Build a dict of `{year: total}` using a dict comprehension and `zip`.

```python
flights = sns.load_dataset('flights')

# Your code here
gg = flights.groupby('year').agg(

    total = ('passengers','mean'),
    peak = ('passengers','max'),
    low = ('passengers','min'),
    seasonal_range = pd.NamedAgg(column='passengers', aggfunc=(lambda x: x.max() - x.min()))

)

print(gg['total'].idxmax(),'has the most total passengers', gg['seasonal_range'].idxmax(),'has the widest seasonal range')

print(gg.index[np.argsort(-gg['seasonal_range'])])

{y: t for y,t in zip(gg.index, gg['total'])}

```

## Level 2 — Sales reps

`reps` records individual sales calls across four regions and four quarters. Before grouping, add `conversion = deals_closed / calls_made`.

Group by `region` and `quarter`. Use named agg to compute `avg_revenue` (mean), `avg_deals` (mean of `deals_closed`), `avg_conversion` (mean of `conversion`), and `n` (count). Then answer: which (`region`, `quarter`) had the highest average conversion rate? Which had the highest average revenue? Is there a correlation between `avg_deals` and `avg_revenue` across all groups? Print the value.

```python
np.random.seed(15)
n = 120
reps = pd.DataFrame({
    'region':       np.random.choice(['North', 'South', 'East', 'West'], n),
    'quarter':      np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], n),
    'calls_made':   np.random.randint(50, 200, n),
    'deals_closed': np.random.randint(5, 40, n),
    'revenue':      np.random.uniform(10_000, 80_000, n).round(-2),
})

# Your code here


reps['conversion'] = reps['deals_closed']/reps['calls_made']

res = reps.groupby(['region','quarter']).agg(
    avg_revenue = ('revenue','mean'),
    avg_deals = ('deals_closed','mean'),
    avg_conversion = ('conversion','mean'),
    n = ('revenue', 'count')
)
print(res)

print(res['avg_conversion'].idxmax(),'has the highest conversion rate')

print(res['avg_revenue'].idxmax(),'has the higheset average revenue')
c = np.corrcoef(res['avg_deals'],res['avg_revenue'])[0,1]
print(f'there is correlation of {c:.2f}')
print('mildly negatively correlated')
```

## Level 3 — Health expenditure

Use `sns.load_dataset('healthexp')`. The dataset has `Year`, `Country`, `Spending_USD`, and `Life_Expectancy` — one row per country per year.

Group by `Country`. Build a named-agg summary with `avg_spending`, `avg_life_exp`, `spending_range` (max minus min of `Spending_USD`), and `n`. No hints.

- Which country had the highest average life expectancy? Which had the most variable spending?
- Across countries, does higher average spending correlate with higher life expectancy? Print the value.
- Rank countries from highest to lowest average spending. Then rank them by life expectancy. Do the two rankings agree? Check with `np.corrcoef` on the rank arrays.

```python
healthexp = sns.load_dataset('healthexp')

# Your code here

g = healthexp.groupby('Country').agg(


    avg_spending = ('Spending_USD','mean'),
    avg_life_exp = ('Life_Expectancy','mean'),
    spending_range = pd.NamedAgg(column='Spending_USD', aggfunc= (lambda x: x.max()- x.min())),
    n = ('Spending_USD','count')
)

print(g['avg_life_exp'].idxmax(),'has the highest average life expectancy')
print(g['spending_range'].idxmax(),'has the most variable spending')
c = np.corrcoef(g['avg_spending'],g['avg_life_exp'])[0,1]
print(f'correlation is {c:.2f}')
print('negatively correlated')

rs = np.argsort(-g['avg_spending'])
rl = np.argsort(-g['avg_life_exp'])


np.corrcoef(rs, rl)[0,1]
print('slightly agree')
```
