### Task 5: Bin & Flag
A DataFrame of loan applicants has been created for you.
1. Use `pd.cut()` to add an `age_group` column: 18–35 → `'Young'`, 35–55 → `'Middle'`, 55+ → `'Senior'`
2. Use `np.clip()` to cap `credit_score` between the 5th and 95th percentile, store as `score_clipped`
3. Use `np.where()` to add a `high_income` column — `True` if `income` > 80000, `False` otherwise

```python
# Starter data — don't change this
np.random.seed(7)
applicants = pd.DataFrame({
    'applicant_id': range(1, 13),
    'age':          np.random.randint(18, 70, size=12),
    'income':       np.random.randint(30000, 150000, size=12),
    'credit_score': np.random.randint(300, 900, size=12),
})

# 1. Add age_group using pd.cut()
applicants['age_group'] = pd.cut(applicants['age'],
                                 bins = [18,35,55,200],
                                 labels= ['Young','Middle','Senior'])

# 2. Clip credit_score between 5th and 95th percentile
p5,p9 = np.percentile(applicants['credit_score'], [5,95])
applicants['score_clipped'] = np.clip(applicants['credit_score'], p5,p9 )

# 3. Add high_income column using np.where()
applicants['high_income'] = np.where(applicants['income']>80000, True, False)
applicants
```

### Task 6: Missing Data + Forward Fill
A time series DataFrame with missing values has been created for you.
1. Count nulls per column
2. Fill missing `temperature` values using `.ffill()` (carry the last known value forward)
3. Fill missing `humidity` values with the `np.nanmedian()`
4. Confirm there are no remaining nulls

```python
# Starter data — don't change this
np.random.seed(3)
readings = pd.DataFrame({
    'hour':        range(1, 13),
    'temperature': [22.1, np.nan, np.nan, 23.5, 24.0, np.nan, 25.1, 25.3, np.nan, 24.8, 24.1, 23.9],
    'humidity':    [55.0, 57.0, np.nan, np.nan, 60.0, 61.5, np.nan, 63.0, 62.0, np.nan, 59.0, 58.0],
})

# 1. Count nulls per column
readings.isnull().sum()

# 2. Fill temperature with forward fill
readings['temperature']= readings['temperature'].ffill()
readings
# 3. Fill humidity with nanmedian

readings['humidity'] = readings['humidity'].fillna(np.nanmedian(readings['humidity']))
# 4. Confirm no remaining nulls

readings.isnull()
```

### Task 7: GroupBy + Transform
A DataFrame of employee salaries has been created for you.

`.groupby().transform()` is different from `.groupby().apply()` — instead of collapsing rows, it returns a value for every row based on its group.

1. Use `.groupby('department').transform('mean')` to add a `dept_avg_salary` column
2. Add a `salary_vs_avg` column = `salary` - `dept_avg_salary`
3. Use `np.where()` to add an `above_avg` column — `True` if `salary_vs_avg` > 0

```python
# Starter data — don't change this
employees = pd.DataFrame({
    'name':       ['Alice', 'Bob', 'Carol', 'Dave', 'Emma', 'Frank', 'Grace', 'Hank'],
    'department': ['Engineering', 'Engineering', 'Marketing', 'Engineering', 'Marketing', 'Sales', 'Sales', 'Marketing'],
    'salary':     [95000, 88000, 62000, 105000, 71000, 54000, 58000, 67000],
})

# 1. Add dept_avg_salary using transform
employees['dept_avg_salary'] = employees.groupby('department')['salary'].transform('mean')
employees
# 2. Add salary_vs_avg column
employees['salary_vs_avg'] = employees['salary']-employees['dept_avg_salary']

# 3. Add above_avg column using np.where()
employees['above_avg'] = np.where(employees['salary_vs_avg']>0, True, False)
```

### Task 8: Lag & Daily Change
A stock price DataFrame has been created for you.
1. Use `.shift(1)` to add a `prev_close` column (previous day's closing price)
2. Add a `daily_change` column = `close` - `prev_close`
3. Add a `pct_change` column = `daily_change` / `prev_close` * 100
4. Add a `rolling_5` column with a 5-day rolling mean of `close`

```python
# Starter data — don't change this
np.random.seed(42)
dates = pd.date_range('2025-01-01', periods=20, freq='B')  # business days
stock = pd.DataFrame({
    'date':  dates,
    'close': np.round(100 + np.cumsum(np.random.normal(0, 2, size=20)), 2),
})
stock = stock.set_index('date')

# 1. Add prev_close using .shift(1)
stock['prev_close'] = stock['close'].shift(1)

# 2. Add daily_change
stock['daily_change'] = stock['close'] - stock['prev_close']

# 3. Add pct_change
stock['pct_change'] = stock['daily_change']/stock['prev_close']*100

# 4. Add 5-day rolling mean of close
stock['rolling_5'] = stock['close'].rolling(5).mean()
stock
```

### Task 10: Full Pipeline
Load the diamonds dataset and run a full analysis:
1. Drop missing values
2. Use `.str.upper()` to standardise the `cut` column
3. Add a `price_per_carat` column = price / carat
4. Use `pd.cut()` to bin `price_per_carat` into: Budget (<3000), Mid (3000–6000), Premium (6000+)
5. Use `.groupby().transform('mean')` to add a `cut_avg_price` column (avg price per cut group)
6. Use `pd.pivot_table()` to show mean `price_per_carat` for each `cut` × `color` combination
7. Find the correlation between `carat` and `price` using `np.corrcoef()`

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv'
diamonds = pd.read_csv(url)

# 1. Drop missing values
diamonds = diamonds.dropna()

# 2. Standardise cut to uppercase
diamonds['cut'] = diamonds['cut'].str.upper()

# 3. Add price_per_carat column
diamonds['price_per_carat'] = diamonds['price']/diamonds['carat']

# 4. Bin price_per_carat into Budget / Mid / Premium
diamonds['group'] = pd.cut(diamonds['price_per_carat'],
                           bins=[0,3000,6000,np.inf],
                           labels=['Budget','Mid','Premium'])

# 5. Add cut_avg_price using transform

diamonds['cut_avg_price'] = diamonds.groupby('cut',observed=True)['price'].transform('mean')
# 6. Pivot table: mean price_per_carat per cut × color
pd.pivot_table(diamonds,
               values='price_per_carat',
               index = 'cut',
               columns= 'color')

# 7. Correlation between carat and price
np.corrcoef(diamonds['carat'], diamonds['price'])
```
