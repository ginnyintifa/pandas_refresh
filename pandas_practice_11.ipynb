### Task 1: DateTime Operations
A DataFrame of project records has been created for you.

1. Convert `start_date` and `end_date` to datetime
2. Add a `duration_days` column = number of days between start and end
3. Extract the `start_month` as a new column
4. Find the mean and standard deviation of `duration_days` using NumPy

```python
# Starter data — don't change this
projects = pd.DataFrame({
    'project':    ['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon', 'Zeta', 'Eta', 'Theta'],
    'start_date': ['2023-01-10', '2023-03-05', '2023-06-20', '2023-08-01',
                   '2023-02-14', '2023-07-30', '2023-04-11', '2023-09-25'],
    'end_date':   ['2023-04-18', '2023-05-22', '2023-09-10', '2023-11-15',
                   '2023-06-01', '2023-12-05', '2023-07-07', '2024-01-30'],
    'team_size':  [4, 7, 3, 10, 5, 8, 6, 2],
})

# Your code here
projects['start_date'] = pd.to_datetime(projects['start_date'])
projects['end_date'] = pd.to_datetime(projects['end_date'])

projects['duration_days'] = (projects['end_date']-projects['start_date']).dt.days
projects['start_month'] = projects['start_date'].dt.month
np.nanmean(projects['duration_days'])
np.nanstd(projects['duration_days'])
```

### Task 2: Handling Missing Data
A DataFrame with missing values has been created for you.

1. Count missing values per column
2. Add a `complete` column — `True` if the row has no missing values, `False` otherwise
3. Fill missing `temperature` values with the column mean
4. Fill missing `condition` values with `'Unknown'`
5. Drop rows where `station` is missing

```python
# Starter data — don't change this
weather = pd.DataFrame({
    'station':     ['A', 'B', np.nan, 'D', 'E', 'F', np.nan, 'H'],
    'temperature': [22.1, np.nan, 18.5, np.nan, 25.3, 19.8, 21.0, np.nan],
    'condition':   ['Sunny', 'Cloudy', 'Rainy', np.nan, 'Sunny', np.nan, 'Cloudy', 'Rainy'],
    'humidity':    [55, 72, 80, 68, 45, np.nan, 60, 75],
})

# Your code here
weather.isna().sum()
weather['complete'] = weather.isna().sum(axis=1)==0
weather['temperature'] = weather['temperature'].fillna(np.nanmean(weather['temperature']))
weather['condition'] = weather['condition'].fillna("Unknown")
weather = weather.dropna(subset='station')
```

### Task 4: String operations and binning
A DataFrame of book listings has been created for you.

1. Convert `title` to uppercase and store as `title_upper`
2. Use `.map()` with a dictionary to add a `genre_code` column: `'Fiction'→'F'`, `'Non-Fiction'→'NF'`, `'Science'→'S'`, `'History'→'H'`
3. Use `pd.cut()` to bin `price` into 3 groups — label them `['Budget', 'Mid', 'Premium']`
4. Use `np.log()` to add a `log_price` column

```python
# Starter data — don't change this
np.random.seed(5)
books = pd.DataFrame({
    'title': ['the great novel', 'history of time', 'python basics', 'world war ii',
              'machine learning', 'ancient rome', 'fiction tales', 'dna decoded'],
    'genre': ['Fiction', 'History', 'Science', 'History',
              'Science', 'History', 'Fiction', 'Science'],
    'price': [8.99, 24.50, 39.99, 15.00, 49.99, 18.75, 11.50, 34.99],
    'rating': np.round(np.random.uniform(3.0, 5.0, 8), 1),
})

# Your code here
books['title_upper'] = books['title'].str.upper()
maps = {'Fiction':'F',
        'Non-Fiction': 'NF',
        'Science': 'S',
        'History': 'H'}
books['genre_code'] = books['genre'].map(maps)
pd.cut(books['price'], bins=3, labels=['Budget','Mid','Premium'])
books['log_price'] = np.log(books['price'])
```

### Task 5: groupby and aggregation
A DataFrame of gym members has been created for you.

1. Group by `membership_type` and find mean and max of `sessions_per_month`
2. Add a `avg_sessions_for_type` column — each row gets the mean sessions for its membership type
3. Find the membership type with the highest average spend using `.idxmax()`
4. Use `np.percentile()` to find the 25th and 75th percentile of `monthly_spend`

```python
# Starter data — don't change this
np.random.seed(9)
gym = pd.DataFrame({
    'member_id':       range(1, 21),
    'membership_type': np.random.choice(['Basic', 'Standard', 'Premium'], size=20),
    'sessions_per_month': np.random.randint(1, 20, size=20),
    'monthly_spend':   np.round(np.random.uniform(20, 120, size=20), 2),
})

# Your code here
gym.groupby('membership_type')['sessions_per_month'].agg(['mean','max'])
gym['avg_sessions_for_type']= gym.groupby('membership_type')['sessions_per_month'].transform('mean')
gym.groupby('membership_type')['monthly_spend'].mean().idxmax()
np.percentile(gym['monthly_spend'], [25,75])
```

### Task 6: resample() with Time Series
A DataFrame of hourly website traffic over 14 days has been created for you.

1. Set `timestamp` as the index
2. Resample to daily total and daily mean `visitors`
3. Add a `cumulative_visitors` column to the daily totals
4. Find the day with the highest total visitors

```python
# Starter data — don't change this
np.random.seed(42)
traffic = pd.DataFrame({
    'timestamp': pd.date_range('2024-03-01', periods=14*24, freq='h'),
    'visitors':  np.random.randint(50, 800, size=14*24),
})

# Your code here
traffic = traffic.set_index('timestamp')
dt = traffic.resample('D').sum()
traffic.resample('D').mean()
dt['cumulative_visitors']=dt['visitors'].cumsum()
dt
dt['visitors'].idxmax()
```

### Task 7: Full Pipeline
Load the tips dataset and run a full analysis:

1. Add a `tip_pct` column = tip as a percentage of total bill
2. Bin `total_bill` into 3 groups — label them `['Low', 'Mid', 'High']`, store as `bill_group`
3. Get mean and std of `tip_pct` per `bill_group`
4. Add a `avg_tip_pct_by_day` column using `transform`
5. Use `np.corrcoef()` to find the correlation between `total_bill` and `tip`
6. Build a pivot table showing mean `tip_pct` for each `day` × `time` combination

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
tips = pd.read_csv(url)

# Your code here

tips['tip_pct'] = tips['tip']/tips['total_bill']
tips['bill_group'] = pd.cut(tips['total_bill'],bins=3, labels=['Low','Mid','High'])
tips.groupby('bill_group',observed=True)['tip_pct'].agg(['mean','std'])
tips['avg_tip_pct_by_day']= tips.groupby('day')['tip_pct'].transform('mean')
np.corrcoef(tips['total_bill'],tips['tip'])
p = tips.pivot_table(index='day',columns='time',values='tip_pct')
p
pd.crosstab(tips['time'],tips['day'])
```
