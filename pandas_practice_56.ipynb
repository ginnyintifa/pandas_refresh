## Level 1 — City climate snapshots

`pd.wide_to_long(df, stubnames, i, j, sep='_', suffix=r'\d+')` reshapes wide data where multiple columns share a prefix into long format.

```python
pd.wide_to_long(df,
    stubnames=['sales', 'cost'],   # column prefixes to fold — each becomes a column
    i=['store', 'region'],         # columns that identify each entity → become the index
    j='year',                      # name for the new column that holds the suffix
    sep='_',                       # separator between stub and suffix
    suffix=r'\d+',                 # regex for the suffix — r'\d+' for numbers, r'\w+' for strings like 'Jan'
)
```

Given `climate` below, where each row is a city and columns hold monthly readings by prefix:

- Reshape into long format. Each row should represent one `(city, month)` pair, with `temp` and `rain` as separate columns.
- Which city has the highest mean temperature across all months?
- Which month has the highest mean rainfall?
- Use `np.corrcoef` to measure the correlation between `temp` and `rain` across all city–month pairs. Is warmer weather wetter or drier here?

````python
climate = pd.DataFrame({
    'city':     ['Austin', 'Denver', 'Miami', 'Portland', 'Chicago'],
    'temp_Jan': [50, 35, 72, 48, 29],
    'temp_Apr': [70, 52, 80, 57, 52],
    'temp_Jul': [95, 88, 91, 75, 84],
    'temp_Oct': [72, 57, 82, 57, 58],
    'rain_Jan': [1.9, 0.5, 1.9, 5.2, 1.8],
    'rain_Apr': [2.1, 1.8, 2.8, 2.6, 3.7],
    'rain_Jul': [1.8, 1.4, 5.8, 0.4, 3.5],
    'rain_Oct': [3.1, 1.1, 6.1, 3.2, 2.8],
})

# Your code here

climate_long = pd.wide_to_long(
    climate,
    stubnames=['temp','rain'],
    i = 'city',
    j = 'month',
    sep = '_',
    suffix = r'\w+'
)

climate_long.groupby(level ='city')['temp'].mean().idxmax()
climate_long.groupby(level ='month')['rain'].mean().idxmax()

np.corrcoef(climate_long['temp'], climate_long['rain'])[0,1]


## the warmer the wetter 
````

## Level 3 — Diamonds

Use pandas and NumPy to answer these questions. No function hints — choose the right tool.

1. Make `cut` an ordered category: `Fair < Good < Very Good < Premium < Ideal`.
2. Within each `color` grade, what fraction of diamonds have a cut of `Very Good` or better? Return a Series sorted from highest to lowest fraction.
3. Is `carat` more strongly correlated with `price`, or `table` more strongly correlated with `price`? Report both correlation coefficients and name the winner.
4. Add `price_per_carat = price / carat`. Use `np.argsort` to rank the `cut` grades from cheapest to most expensive per carat on average.

````python
diamonds = sns.load_dataset('diamonds')

# Your code here
co = pd.CategoricalDtype(['Fair','Good','Very Good','Premium','Ideal'], ordered=True)
diamonds['cut'] = diamonds['cut'].astype(co)
s = pd.Series([(diamonds.loc[diamonds['color']==c,'cut']>='Very Good').mean() for c in diamonds['color'].unique()])
s.index = diamonds['color'].unique()
s = s.sort_values(ascending = False)
cc = np.corrcoef(diamonds['carat'], diamonds['price'])[0,1]
tc = np.corrcoef(diamonds['table'], diamonds['price'])[0,1]
cdict = {'carat_cor': cc,
 'table_cor': tc}
print(cdict, 'winner is: ', max(cdict, key = cdict.get))

diamonds['price_per_carat'] = diamonds['price']/diamonds['carat']
g = diamonds.groupby('cut', observed=True)['price_per_carat'].mean()

np.argsort(g)

````
