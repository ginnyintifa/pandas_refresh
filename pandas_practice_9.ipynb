### Task 3: apply() with a Custom Function
A DataFrame of exam scores has been created for you.
1. Write a function `letter_grade(score)` that returns `'A'` (≥90), `'B'` (≥80), `'C'` (≥70), `'D'` (≥60), or `'F'` (below 60)
2. Use `.apply()` to add a `grade` column
3. Add a `z_score` column: `(score - mean) / std` — use `np.mean()` and `np.std()`
4. Use `.value_counts()` to count how many students got each grade

```python
# Starter data — don't change this
np.random.seed(1)
students = pd.DataFrame({
    'name':  ['Alice', 'Bob', 'Carol', 'Dave', 'Emma', 'Frank', 'Grace', 'Hank', 'Ivy', 'Jack'],
    'score': np.random.randint(50, 100, size=10),
})

# Your code here
def letter_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
    
students['grade'] = students['score'].apply(lambda x: letter_grade(x))

students['z_score'] = (students['score'] - np.mean(students['score']))/np.std(students['score'])

students['grade'].value_counts()
```

### Task 4: Reshape with `.melt()`
A wide-format DataFrame of monthly sales has been created for you.

`.melt()` converts wide format (one column per month) to long format (one row per month).

1. Use `.melt()` to reshape so each row is one `(region, month, sales)` observation. Set `id_vars='region'`, `var_name='month'`, `value_name='sales'`
2. Sort the result by `region`, then `month`

### Task 5: Multi-Column GroupBy + .agg()
A DataFrame of sales transactions has been created for you.

`.agg()` lets you apply multiple aggregation functions at once.

1. Group by `region` and `category`, then use `.agg()` to compute `min`, `max`, and `mean` of `amount` in one step
2. Group by `region` only and find the total `amount` and number of transactions (`count`) per region
3. Find which `region` + `category` combination has the highest mean `amount`

```python
# Starter data — don't change this
np.random.seed(5)
transactions = pd.DataFrame({
    'region':   np.random.choice(['North', 'South', 'East'], size=20),
    'category': np.random.choice(['Electronics', 'Clothing', 'Food'], size=20),
    'amount':   np.random.randint(20, 500, size=20),
})

# Your code here

transactions.groupby(['region','category'])['amount'].agg(['min','max','mean'])
transactions.groupby('region')['amount'].agg(['sum','count'])
#transactions.groupby(['region','category'])['amount'].mean().idxmax()
```

### Task 6: Cumulative & Rolling Stats
A DataFrame of daily website visitors has been created for you.
1. Add a `cumulative` column using `.cumsum()` — running total of visitors
2. Add a `rolling_7` column — 7-day rolling mean of `visitors`
3. Add a `rolling_std` column — 7-day rolling standard deviation
4. Use `np.argmax()` to find the index of the day with the highest visitor count, then print that row

```python
# Starter data — don't change this
np.random.seed(9)
traffic = pd.DataFrame({
    'date':     pd.date_range('2025-01-01', periods=30, freq='D'),
    'visitors': np.random.randint(800, 3000, size=30),
})
traffic = traffic.set_index('date')

# Your code here
traffic['cumulative'] = traffic['visitors'].cumsum()
traffic['rolling_7'] = traffic['visitors'].rolling(7).mean()
traffic['rolling_std'] = traffic['visitors'].rolling(7).std()

traffic.iloc[np.argmax(traffic['visitors'])]
```

### Task 7: Full Pipeline
Load the tips dataset and run a full analysis:
1. Drop missing values
2. Add a `tip_pct` column = `tip / total_bill * 100`
3. Use `pd.cut()` to bin `tip_pct` into: Low (<10), Fair (10–20), Generous (20+)
4. Use `.groupby('day').agg()` to get the mean and std of `tip_pct` per day
5. Use `.groupby('day').transform('mean')` to add a `day_avg_tip_pct` column
6. Use `np.corrcoef()` to find the correlation between `total_bill` and `tip`
7. Use `pd.pivot_table()` to show mean `tip_pct` for each `day` × `time` combination

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
tips = pd.read_csv(url)

# Your code here
tips = tips.dropna()
tips['tip_pct'] = tips['tip']/tips['total_bill']*100
pd.cut(tips['tip_pct'],bins=[0,10,20,np.inf], labels = ['Low','Fair','Generous'])
tips.groupby('day')['tip_pct'].agg(['mean','std'])
tips['day_avg_tip_pct'] = tips.groupby('day')['tip_pct'].transform('mean')
np.corrcoef(tips['total_bill'],tips['tip'])
pd.pivot_table(tips,
               index='day',
               columns='time',
               values='tip_pct',
               aggfunc='mean')
```
