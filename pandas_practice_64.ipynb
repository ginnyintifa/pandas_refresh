## Level 1 — Students

`students` records exam scores and weekly study hours across five subjects.

Group by `subject`. Build a named-agg summary with `n`, `avg_score`, `avg_hours`, and `score_range` (max minus min of `score`). Then:

- Which subject has the widest score range?
- Across subjects, do more study hours tend to go with higher scores? Check with `np.corrcoef`.
- Build a dict mapping each subject to its `avg_score` using `zip` and `dict()`.

```python
np.random.seed(42)
subjects = ['Math', 'English', 'Science', 'History', 'Art']
students = pd.DataFrame({
    'student_id':    np.arange(1, 61),
    'subject':       np.repeat(subjects, 12),
    'score':         np.clip(np.random.normal(np.repeat([78,72,81,70,85], 12), 12, 60).astype(int), 0, 100),
    'hours_studied': np.clip(np.random.normal(np.repeat([5,4,6,3,4], 12), 1.5, 60).round(1), 0.5, 10)})

# Your code here

gs = students.groupby('subject').agg(
    n = ('student_id','count'),
    avg_score = ('score','mean'),
    avg_hours = ('hours_studied','mean'),
    score_range = pd.NamedAgg(column='score', aggfunc= lambda x: x.max() - x.min())
)

print(gs)

print(gs['score_range'].idxmax(),'has the widest score range')

np.corrcoef(gs['avg_hours'],gs['avg_score'])[0,1]
print('positively correlated')

{s:avs for s, avs in zip(gs.index, gs['avg_score'])}
```

## Level 3 — MPG

Use `sns.load_dataset('mpg').dropna()`. Group by `origin` and `cylinders` (`observed=True`).

Build a named-agg summary with `avg_mpg`, `avg_weight`, and `n`. Then answer these questions — no hints, no sub-steps:

- Which (`origin`, `cylinders`) combination is the most fuel-efficient? Which is the least?
- Within each origin, what is the range of `avg_mpg` across cylinder counts? Which origin has the widest spread? Do this without a loop.
- Across all groups, is heavier average weight associated with lower mpg? Quantify it.

```python
mpg = sns.load_dataset('mpg').dropna()

# Your code here

g = mpg.groupby(['origin','cylinders'], observed = True).agg(
    avg_mpg = ('mpg','mean'),
    avg_weight = ('weight','mean'),
    n = ('mpg','count')
)

print(g['avg_mpg'].idxmax(),'is the most fuel efficient')
print(g['avg_mpg'].idxmin(),'is the least fuel efficient')

ug = g['avg_mpg'].unstack('cylinders')
rg = np.nanmax(ug,axis=1) - np.nanmin(ug, axis = 1)
print(ug.index[np.argsort(-rg)][0],'has the widest spread')

np.corrcoef(-g['avg_mpg'],g['avg_weight'])[0,1]
print('yes')
```
