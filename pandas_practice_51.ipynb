## Level 1 — flight connections

A flight operations team tracks international routes. Answer these five questions:

1. What is the **local departure time** at each origin city?
2. What is the **local arrival time** at each destination city?
3. How long is each flight in **hours** (float, 1 decimal)?
4. Flag `is_red_eye`: True where local departure hour is ≥ 22 or < 6.
5. Which route has the **longest flight time**?

Hint: localize `depart_utc` / `arrive_utc` to UTC first, then `.apply()` each row to its timezone.

```python
flights = pd.DataFrame({
    'flight':     ['AA101', 'LH202', 'JL303', 'BA404', 'QF505'],
    'origin_tz':  ['US/Pacific', 'Europe/Berlin', 'Asia/Tokyo', 'Europe/London', 'Australia/Sydney'],
    'dest_tz':    ['Europe/London', 'US/Eastern', 'US/Pacific', 'Asia/Dubai', 'US/Pacific'],
    'depart_utc': pd.to_datetime(['2024-06-10 06:00', '2024-06-10 06:00', '2024-06-10 15:00',
                                   '2024-06-10 09:00', '2024-06-10 11:30']),
    'arrive_utc': pd.to_datetime(['2024-06-11 00:00', '2024-06-10 14:30', '2024-06-11 05:00',
                                   '2024-06-10 21:00', '2024-06-11 01:30']),
})

# Your code here
flights['depart_utc'] = flights['depart_utc'].dt.tz_localize('UTC')
flights['local_departure_time'] = flights.apply(lambda row: row['depart_utc'].tz_convert(row['origin_tz']), axis = 1)
flights['arrive_utc'] = flights['arrive_utc'].dt.tz_localize('UTC')
flights['local_arrival_time'] = flights.apply(lambda row: row['arrive_utc'].tz_convert(row['dest_tz']), axis = 1)
fh = round((flights['arrive_utc'] -flights['depart_utc']).dt.total_seconds()/3600,1)
print(fh)
### need to do element wise because on different time zone 
flights['is_red_eye'] = flights['local_departure_time'].apply(lambda x: (x.hour >= 22) | (x.hour <6) )

flights.loc[fh.idxmax()]['flight']

```

## Level 2 — subscription billing

Monthly SaaS subscription data. No step-by-step — figure out the approach.

- `billing_date`: 1 month after `signup`. `contract_end`: 12 months after `signup`.
- `is_expiring_soon`: contract ends within 3 months of 2026-04-18.
- Convert `plan` to `category`.
- Total revenue and subscriber count per `plan`.
- Plain Python: build a dict mapping each plan name → list of `subscriber_id`s in that plan.

```python
subs = pd.DataFrame({
    'subscriber_id': ['S001', 'S002', 'S003', 'S004', 'S005', 'S006', 'S007'],
    'signup':        pd.to_datetime(['2025-10-01', '2025-11-15', '2026-01-03', '2025-08-20',
                                     '2026-02-28', '2025-12-01', '2026-03-10']),
    'plan':          ['basic', 'pro', 'enterprise', 'basic', 'pro', 'enterprise', 'basic'],
    'monthly_fee':   [9.99, 29.99, 99.99, 9.99, 29.99, 99.99, 9.99],
})

# Your code here

subs['billing_date'] = subs['signup']+ pd.DateOffset(months=1)
subs['contract_end'] = subs['signup'] + pd.DateOffset(months = 12)
#subs['is_expiring_soon'] = (subs['contract_end'] - pd.to_datetime('2026-04-18')).dt.days <= 3*31
subs['is_expiring_soon']  = subs['contract_end'] <= pd.to_datetime('2026-04-18') + pd.DateOffset(months=3)
subs['plan'] = subs['plan'].astype('category')

subs.groupby('plan', observed=True)['subscriber_id'].count()
subs['rev'] = subs['monthly_fee']*12
subs.groupby('plan', observed = True)['rev'].sum()

```

```python
results = {}
for plan, sid in zip(subs['plan'], subs['subscriber_id']):
    if plan not in results:
        results[plan] = []
    results[plan].append(sid)
```

## Level 3 — warehouse sensor pipeline

Raw temperature readings from warehouse sensors. Build a pipeline using `.pipe()`.

Write these three functions and chain them on `readings.copy()`:

- **`standardize(df)`** — parse `recorded_at` as datetime, localize to UTC. Extract `sensor_id` from `tag` (format: `WHSE-X/SNS-NN`) using `.str.extract()`. Drop `tag`.
- **`enrich(df)`** — convert `sensor_id` to `category`. Add `local_time`: `recorded_at` converted to `US/Eastern`. Flag `is_business_hours`: local hour is 8–17 inclusive.
- **`flag_anomalies(df)`** — add `rolling_avg`: 3-reading rolling mean of `temp_c` **per sensor** using `.groupby().transform()`. Add `is_anomaly`: True where `temp_c` deviates from `rolling_avg` by more than 2°C.

After the chain:
- Mean temperature per sensor.
- Anomaly count during business hours vs. outside.
- Which sensor has the highest anomaly rate?

```python
readings = pd.DataFrame({
    'tag':         ['WHSE-A/SNS-01', 'WHSE-B/SNS-02', 'WHSE-A/SNS-01', 'WHSE-C/SNS-03',
                    'WHSE-B/SNS-02', 'WHSE-A/SNS-01', 'WHSE-C/SNS-03', 'WHSE-B/SNS-02',
                    'WHSE-A/SNS-01', 'WHSE-C/SNS-03', 'WHSE-B/SNS-02', 'WHSE-C/SNS-03'],
    'recorded_at': ['2024-03-20 13:00', '2024-03-20 13:05', '2024-03-20 14:00',
                    '2024-03-20 14:00', '2024-03-20 15:00', '2024-03-20 22:00',
                    '2024-03-20 22:30', '2024-03-20 23:00', '2024-03-21 01:00',
                    '2024-03-21 01:30', '2024-03-21 08:00', '2024-03-21 08:30'],
    'temp_c':      [22.1, 21.8, 22.3, 19.5, 22.0, 28.7, 27.0, 21.9, 22.5, 19.8, 22.1, 25.9],
})

# Your code here
def standardize(df):
    df['recorded_at'] = pd.to_datetime(df['recorded_at']).dt.tz_localize('UTC')
    #df['sensor_id'] = df['tag'].str.extract(r'WHSE-(\w)') + df['tag'].str.extract(r'/SNS-(\d{2})')
    parts = df['tag'].str.extract(r'WHSE-(\w)/SNS-(\d{2})')
    df['sensor_id'] = parts[0]+parts[1]
    df = df.drop(columns = 'tag')
    return df


def enrich(df):
    df['sensor_id'] = df['sensor_id'].astype("category")
    df['local_time'] = df['recorded_at'].dt.tz_convert('US/Eastern')
    df['is_business_hours'] = (df['local_time'].dt.hour>=8) & (df['local_time'].dt.hour<=17)
    return df


def flag_anomalies(df):
    df['rolling_avg'] = df.groupby('sensor_id', observed = True)['temp_c'].transform(lambda x: x.rolling(3).mean())
    df['is_anomaly'] = abs(df['temp_c'] - df['rolling_avg']) >2
    return df 










```

```python
res = readings.copy().pipe(standardize).pipe(enrich).pipe(flag_anomalies)

res.groupby('sensor_id', observed=True)['temp_c'].mean()
res.loc[res['is_anomaly']]['is_business_hours'].value_counts()

res.groupby('sensor_id', observed=True)['is_anomaly'].mean().idxmax()
```
