### Task 5: Aggregations Along Axes
A 2D array of exam scores has been created (rows = students, columns = subjects).
1. Find the **overall** mean score
2. Find the **mean score per student** (mean across columns, `axis=1`)
3. Find the **mean score per subject** (mean across rows, `axis=0`)
4. Find the highest score overall using `np.max()`

```python
# Starter data — don't change this
# Rows = students, Columns = [Math, English, Science]
scores = np.array([[85, 92, 78],
                   [70, 88, 95],
                   [60, 74, 80],
                   [90, 85, 88]])

# 1. Overall mean
scores.mean()

# 2. Mean per student (axis=1)
scores.mean(1)

# 3. Mean per subject (axis=0)
scores.mean(0)

# 4. Highest score overall
scores.max()
```

### Task 6: Boolean Masking

2. Use the mask to extract all scores above 80 as a 1D array
3. Count how many scores are above 80 (hint: `True` counts as 1)
4. Replace all scores below 70 with 70 (curve the grades)

### Task 7: Reshape & Transpose
1. Create a 1D array of integers from 1 to 12 using `np.arange()`
2. Reshape it into a 3×4 matrix using `.reshape()`
3. Transpose it using `.T`
4. Flatten the transposed array back to 1D using `.flatten()`

```python
# 1. 1D array from 1 to 12

p = np.arange(1,13)
# 2. Reshape to 3x4

p2 = p.reshape([3,4])
# 3. Transpose

p2t = p2.T
# 4. Flatten back to 1D
p2tf = p2t.flatten()
```
