### Task 5: Handle Missing Data
A DataFrame with `NaN` values has been created for you.
1. Find which cells are null using `.isnull()`
2. Drop rows with any null values using `.dropna()`
3. Fill null values with the mean of each column using `.fillna()`

```python
# Starter data — don't change this
employees = pd.DataFrame({
    'name':       ['Alice', 'Bob', 'Carol', 'Dave', 'Emma'],
    'salary':     [70000, np.nan, 85000, np.nan, 92000],
    'experience': [3, 5, np.nan, 8, 2],
    'score':      [np.nan, 88, 91, 74, np.nan],
})

# 1. Show which values are null
employees.isnull()

# 2. Drop rows with any null
employees.dropna()

# 3. Fill nulls with the mean of each column
employees.fillna(employees.mean(numeric_only=True))
```

### Task 6: Sort & Rank
A sales DataFrame has been created for you.
1. Sort by `revenue` descending
2. Add a `rank` column (1 = highest revenue) using `.rank()`

```python
# Starter data — don't change this
sales = pd.DataFrame({
    'rep':     ['Alice', 'Bob', 'Carol', 'Dave', 'Emma'],
    'region':  ['North', 'South', 'East', 'West', 'North'],
    'revenue': [42000, 31500, 67800, 55200, 48900],
})

# 1. Sort by revenue descending
sales.sort_values('revenue', ascending=False)

# 2. Add a rank column (1 = highest revenue)
sales['rank'] = sales['revenue'].rank(ascending=False)
sales
```

## Level 3 — Aggregation

### Task 7: GroupBy
Using the `orders` DataFrame below, find the **total revenue per region** using `.groupby()`.

```python
# Starter data — don't change this
orders = pd.DataFrame({
    'region':   ['East', 'West', 'East', 'North', 'West', 'North', 'East'],
    'rep':      ['Alice', 'Bob', 'Carol', 'Dave', 'Alice', 'Bob', 'Dave'],
    'revenue':  [1200, 4500, 800, 3200, 2100, 1750, 950],
})

# Total revenue per region
total_by_region = orders.groupby('region')['revenue'].sum()
total_by_region
```

### Task 8: Pivot Table
Using the same `orders` DataFrame, create a pivot table showing the **total revenue** for each combination of `region` and `rep`.

```python
# Create a pivot table
pivot = pd.pivot_table(
    orders,
    values= 'revenue',
    index= 'region',
    columns='rep',
    aggfunc=sum
)
pivot
```

### Task 9: Merge DataFrames
Two DataFrames are provided. Merge them on `emp_id` so each row has the employee name and their department.

Try `how='inner'`, `how='left'`, and `how='outer'` — what changes?

```python
# Starter data — don't change this
emp_info = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name':   ['Alice', 'Bob', 'Carol', 'Dave'],
})

departments = pd.DataFrame({
    'emp_id':     [1, 2, 3, 5],   # note: id 4 is missing, id 5 is extra
    'department': ['Engineering', 'Marketing', 'Engineering', 'HR'],
})

# Merge on emp_id
merged = pd.merge(emp_info, departments, on='emp_id', how='inner')
merged
```

### Task 10: String Operations
A customer DataFrame has been created for you. Using pandas string methods (`.str`):
1. Convert all `name` values to uppercase
2. Extract the domain from each `email` (everything after the `@`)
3. Filter to only customers whose email is from `gmail.com`

```python
# Starter data — don't change this
customers = pd.DataFrame({
    'name':  ['alice smith', 'bob jones', 'carol white', 'dave brown', 'emma green'],
    'email': ['alice@gmail.com', 'bob@yahoo.com', 'carol@gmail.com', 'dave@outlook.com', 'emma@gmail.com'],
    'age':   [28, 34, 22, 45, 31],
})

# 1. Uppercase names
customers['name'] = customers['name'].str.upper()
customers
# 2. Extract email domain (hint: .str.split('@') or .str.extract())
customers['domain'] = customers['email'].str.split('@').str[1]

# 3. Filter to gmail customers only
gmail_customers = customers[customers['domain'] == "gmail.com"]
gmail_customers
```
